# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

## Изучение данных из файла

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/datasets/real_estate_data.csv', sep = '\t')
display(df.head())
df.info()
df.duplicated().sum()
```

### Вывод

По изучений данных было выявлено, что в данных отсутствуют значения в большей части столбцов. Необходимо будет подробнее рассмотреть каждый из них и принять решение о дальнейшей обработке. Также, после изучения данных методом info() появилось предположение, что в некоторых столбцах не верно указан тип данных.

## Предобработка данных

```python
#переименовываем столбцы
df = df.rename(columns = {'cityCenters_nearest':'city_centers_nearest'})
```

### Столбец 'is_apartment'

```python
#изучаем пропуски в столбце 'is_apartment'
print(df['is_apartment'].head(10))
print(df['is_apartment'].value_counts())

#можно предположить, что заполнены только строки, где значение положительное (только значения True).
#поэтому заполняю пропуски значениями False
df['is_apartment'] = df['is_apartment'].fillna(False) 

#Во время анализа было  выявлено, что данные в столбце типа object, а значения подразумевают тип bool
#Выполним замену типа данных
df['is_apartment'] = df['is_apartment'].astype('bool')
print(df['is_apartment'].head())
```

### Столбец 'ceiling_height'

```python
#изучаем значения в столбце ceiling_height
print(df['ceiling_height'].describe())

#проверяем на наличие пропусков
print(df['ceiling_height'].isna().sum())
df['ceiling_height'].hist(bins=100, range = (2.4,6))
plt.show()

#в данных обнаружено 9118 пропусков.
#значение ниже 2.4 я считаю аномальными. Эти значения для подбора типа замены данных использовать нельзя
#значения выше 6 метров так же считаю аномальными.
seg = df.query('2.4 <= ceiling_height <= 6')
print(seg['ceiling_height'].describe())

#для заполнения отсутствующих данных используем медиану (2.65). 
df['ceiling_height'] = df['ceiling_height'].fillna(seg['ceiling_height'].median())

#print(df['ceiling_height'].describe())
```

### Столбец 'floors_total'

```python
#изучаем пропуски в столбце 'floors_total'
print(df['floors_total'].isna().sum())

#так как пропущенных значений всего 86 (0.4% от данных) строк, 
#и невозможно предположить логичную замену, данные строки удалим из датафрейма
df = df.dropna(axis='index', how='any', subset=['floors_total'])

#меняет тип данных на int, так как этаж обозначается только целым числом
df['floors_total'] = df['floors_total'].astype('int')
```

### Столбец 'balcony'

```python
#изучаем пропуски в столбце 'balcony'
print(df['balcony'].value_counts())
print(df['balcony'].head(10))

#В данных присутствуют значения NaN
#Для них можно предположить логичную замену.
#если человек не указал число балконов — скорее всего, их нет. Выполняем замену значений
df['balcony'] = df['balcony'].fillna(0)
print(df['balcony'].head(10))

#Выполним замену типа данных, так как количество балконов обозначается только целым числом
df['balcony'] = df['balcony'].astype('int')
```

### Столбец 'kitchen_area'

```python
#изучаем значения в столбце kitchen_area
print(df['kitchen_area'].describe())

#проверяем на наличие пропусков
print(df['kitchen_area'].isna().sum())

#так как разброс данных достаточно большой, принято решение заполнить пропуски медианным значением.
#оно близко со средним, но лучше подойдет для этих данных
df['kitchen_area'] = df['kitchen_area'].fillna(df['kitchen_area'].median())

```

### Столбец 'city_centers_nearest'

```python
#изучаем значения в столбце city_centers_nearest
print(df['city_centers_nearest'].isna().sum())
print(df['city_centers_nearest'].describe())
df['city_centers_nearest'].hist(bins=100, range = (181,65968))

#так как разброс данных достаточно большой, принято решение заполнить пропуски медианным значением.
df['city_centers_nearest'] = df['city_centers_nearest'].fillna(df['city_centers_nearest'].median())
```

### Столбец 'locality_name'

```python
#изучаем значения в столбце locality_name
df['locality_name'].describe()
df['locality_name'].isna().sum()

#так как сопоставить расстояние до центра с остальными городами практически невозможно, 
#принято решение удалить строки с пустыми значениями (43 строки)
df = df.dropna(axis='index', how='any', subset=['locality_name'])
```

### Столбец 'living_area'

```python
#изучаем значения в столбце locality_name
print(df['living_area'].describe())
print(df['living_area'].isna().sum())

#так как разброс данных достаточно большой, принято решение заполнить пропуски медианным значением.
df['living_area'] = df['living_area'].fillna(df['living_area'].median())
```

### Вывод


Как я предпологала в выводе первого пункта, в некоторых столбцах был не верно указан тип данных. После более детального изучения каждого столбца я смогла выявить все подобные отклонения, и заменить тип данных на нужный.
Так же в данных отсутствовали значения в большей части столбцов. В идеале необходимо вернуться к заказчику с данным вопросом, и просить выгрузить данные повтороно, что бы заполнить все строки значениями, либо хотя бы уменьшить количество этих пропусков.
Но так как этой возможности не было, я приняла решение заполнять отсутствующие данные медианными значениями. Но были случаи, когда пропущенных строк было не так много, по этому их было проще удалить, чем заполнять.
Такая обработка данных все еще не идеальна, так как очень много значений были созданы искуственно мной и могут не соответствовать реальным данным. Но в данных условиях это лучший возможный вариант.

## Расчёты и добавление результатов в таблицу

```python
#рассчитываем цену квадратного метра
df['price_per_meter'] = df['last_price']/df['total_area']
#округлим значение данного показателя, так как его используют без знаков после запятой
df['price_per_meter'] = round(df['price_per_meter'])
print(df['price_per_meter'].head())
```

```python
#добавляем день недели, месяц и год публикации объявления
df['publication_time'] = pd.to_datetime(df['first_day_exposition'])
df['publication_weekday'] = df['publication_time'].dt.weekday
df['publication_month'] = pd.DatetimeIndex(df['publication_time']).month
df['publication_year'] = pd.DatetimeIndex(df['publication_time']).year
```

```python
#добавляем этаж квартиры; варианты — первый, последний, другой
def floor_count(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
            return 'первый'
    elif floor == floors_total:
            return 'последний'
    return 'другой'
    
df['ap_floor'] = df.apply(floor_count, axis = 1)
    
print(df['ap_floor'].value_counts())
```

```python
#рассчитываем соотношение жилой и общей площади
df['living_to_total'] = df['living_area'] / df['total_area']
print(df['living_to_total'])
```

```python
#рассчитываем соотношение площади кухни к общей
df['kitchen_to_total'] = df['kitchen_area'] / df['total_area']
print(df['kitchen_to_total'])
```
